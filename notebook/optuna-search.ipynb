```python
import optuna
import numpy as np
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_classification

# Sample data generation (use your embeddings data here)

def objective(trial):
    # Define the search space for the SVC hyperparameters
    svc_c = trial.suggest_loguniform('svc_c', 1e-6, 1e2)
    svc_kernel = trial.suggest_categorical('svc_kernel', ['linear', 'poly', 'rbf', 'sigmoid'])
    svc_gamma = trial.suggest_loguniform('svc_gamma', 1e-6, 1e1)
    
    # Create a pipeline
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('svc', SVC(C=svc_c, kernel=svc_kernel, gamma=svc_gamma))
    ])
    
    # Perform cross-validation
    scores = cross_val_score(pipeline, data, df['Успешный результат'], cv=5, scoring='precision')
    
    # Return the mean accuracy
    return scores.mean()

# Create an Optuna study
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=200)

# Print the best trial
print("Best trial:")
trial = study.best_trial
print(f"  Value: {trial.value}")
print("  Params: ")
for key, value in trial.params.items():
    print(f"    {key}: {value}")

# Train final model using the best hyperparameters
```
